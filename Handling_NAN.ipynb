```python
import pandas as pd
iris = pd.read_csv('C:/Users/5530/OneDrive/Desktop/Data_Analytics/Data-Analytics/Terrorist_data/terrorismData.csv')
df = iris.copy()
```

```python
df["Gender"] = 'Female'

df
print(len(df.columns))
```

```python
df.loc[0:9, 'Gender'] ='Male'
# df.head(10)
def f(s):
  if s=='Male':
    return 0
  else:
    return 1

df['Sex'] = df.Gender.apply(f)
df.head()
```

```python
mean_value = df[['Killed','Wounded']].mean()
df[['Killed','Wounded']]=df[['Killed','Wounded']].astype(int).fillna(mean_value[['Killed','Wounded']])
```
